```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Load and prepare data
df = pd.read_csv('Employee_Salary_Dataset.csv')
df.head()
```

```python
# Advanced feature engineering
df['Experience_Age_Ratio'] = df['Experience_Years'] / (df['Age'] + 1e-5)
df['Performance_Experience'] = df['Performance_Rating'] * df['Experience_Years']
df['Seniority'] = np.where(df['Experience_Years'] > 15, 1, 0)
df['Education_Level'] = df['Education'].map({
    'High School': 0,
    'Bachelor': 1,
    'Master': 2,
    'PhD': 3
})
```

```python
# Handle outliers - winsorization
q1 = df['Salary'].quantile(0.05)
q3 = df['Salary'].quantile(0.95)
df['Salary'] = np.where(df['Salary'] < q1, q1, df['Salary'])
df['Salary'] = np.where(df['Salary'] > q3, q3, df['Salary'])
```

```python
# Prepare data
X = df.drop(['ID', 'Salary'], axis=1)
y = df['Salary']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# Preprocessing setup
num_features = ['Experience_Years', 'Age', 'Performance_Rating', 
                'Experience_Age_Ratio', 'Performance_Experience', 'Education_Level']
cat_features = ['Gender', 'Department', 'Location', 'Seniority']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ])
```

```python
# Model definitions
models = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42, eval_metric='rmse'),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "ElasticNet": ElasticNet(random_state=42),
    "Ridge": Ridge(random_state=42)
}
```

```python
# Hyperparameter grids
param_grids = {
    "Random Forest": {
        'regressor__n_estimators': [400, 500],
        'regressor__max_depth': [25, 30, None],
        'regressor__min_samples_split': [5, 10],
        'regressor__max_features': [0.7, 'sqrt']
    },
    "XGBoost": {
        'regressor__n_estimators': [600, 700],
        'regressor__learning_rate': [0.05, 0.1],
        'regressor__max_depth': [8, 10],
        'regressor__colsample_bytree': [0.7, 0.9],
        'regressor__gamma': [0, 0.1]
    },
    "Gradient Boosting": {
        'regressor__n_estimators': [500, 600],
        'regressor__learning_rate': [0.05, 0.1],
        'regressor__max_depth': [6, 8],
        'regressor__subsample': [0.8, 1.0],
        'regressor__max_features': ['sqrt', 0.7]
    },
    "ElasticNet": {
        'regressor__alpha': [0.001, 0.01, 0.1],
        'regressor__l1_ratio': [0.3, 0.5, 0.7],
        'regressor__max_iter': [5000]
    },
    "Ridge": {
        'regressor__alpha': [0.1, 1.0, 10.0],
        'regressor__max_iter': [5000]
    }
}
```

```python

# Model training and evaluation
best_model = None
best_score = 0
results = []

for name, model in models.items():
    # Create pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])
    
    # Grid search
    grid = GridSearchCV(
        pipeline,
        param_grids[name],
        cv=5,
        scoring='r2',
        n_jobs=-1,
        verbose=1
    )
    grid.fit(X_train, y_train)
    
    # Best model from grid search
    model = grid.best_estimator_
    cv_score = grid.best_score_
    
    # Evaluate on test set
    y_pred = model.predict(X_test)
    test_r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    
    # Store results
    results.append({
        'Model': name,
        'CV R²': cv_score,
        'Test R²': test_r2,
        'RMSE': rmse,
        'MAE': mae,
        'Best Params': grid.best_params_
    })
    
    print(f"\n{name} Performance:")
    print(f"CV R²: {cv_score:.4f}")
    print(f"Test R²: {test_r2:.4f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    
    # Update best model
    if test_r2 > best_score:
        best_score = test_r2
        best_model = model

# Results summary
results_df = pd.DataFrame(results)
print("\n" + "="*50)
print("MODEL PERFORMANCE SUMMARY")
print("="*50)
print(results_df.sort_values('Test R²', ascending=False))
```

```python
# Best model evaluation
y_pred = best_model.predict(X_test)
residuals = y_test - y_pred

print("\n" + "="*50)
print("BEST MODEL PERFORMANCE")
print("="*50)
print(f"Model Type: {type(best_model.named_steps['regressor']).__name__}")
print(f"Test R²: {best_score:.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred):.2f}")
```

```python
# Feature Importance Visualization
if hasattr(best_model.named_steps['regressor'], 'feature_importances_'):
    # Get feature names
    cat_encoder = best_model.named_steps['preprocessor'].named_transformers_['cat']
    cat_features = cat_encoder.get_feature_names_out(cat_features)
    all_features = num_features + list(cat_features)
    
    importances = best_model.named_steps['regressor'].feature_importances_
    indices = np.argsort(importances)[-15:]
    
    plt.figure(figsize=(12, 8))
    plt.barh(range(len(indices)), importances[indices], align='center')
    plt.yticks(range(len(indices)), [all_features[i] for i in indices])
    plt.title('Top 15 Feature Importances')
    plt.xlabel('Relative Importance')
    plt.tight_layout()
    plt.savefig('feature_importances.png', dpi=300)
    plt.show()
```

```python
# Actual vs Predicted Plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.title('Actual vs Predicted Salaries', fontsize=14)
plt.xlabel('Actual Salary ($)', fontsize=12)
plt.ylabel('Predicted Salary ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('actual_vs_predicted.png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_squared_error
from scipy import stats
import scipy.stats as stats
import matplotlib.gridspec as gridspec

# 1. Error Distribution Comparison
plt.figure(figsize=(12, 8))
residuals = y_test - y_pred
abs_errors = np.abs(residuals)

```

```python
plt.subplot(2, 1, 1)
sns.kdeplot(residuals, fill=True, color='royalblue')
plt.axvline(x=0, color='r', linestyle='--')
plt.title('Error Distribution Analysis', fontsize=16)
plt.xlabel('Prediction Error ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(alpha=0.3)

plt.subplot(2, 1, 2)
sns.boxplot(x=abs_errors, color='lightgreen')
plt.title('Absolute Error Distribution', fontsize=14)
plt.xlabel('Absolute Prediction Error ($)', fontsize=12)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('error_distribution.png', dpi=300)
plt.show()

```

```python
# 2. Cumulative Accuracy Plot
plt.figure(figsize=(12, 8))
error_tolerance = np.linspace(0, 100000, 100)
accuracy = [np.mean(abs_errors <= tol) for tol in error_tolerance]

plt.plot(error_tolerance, accuracy, 'b-', linewidth=2.5)
plt.fill_between(error_tolerance, accuracy, alpha=0.2, color='blue')
plt.axhline(y=0.9, color='r', linestyle='--', alpha=0.7)
plt.axvline(x=20000, color='g', linestyle='--', alpha=0.7)
plt.title('Cumulative Prediction Accuracy', fontsize=16)
plt.xlabel('Acceptable Error Threshold ($)', fontsize=12)
plt.ylabel('Percentage of Predictions', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(['Accuracy', '90% Accuracy Level', '$20k Error Threshold'])
plt.savefig('cumulative_accuracy.png', dpi=300)
plt.show()

```

```python
# 3. Error by Experience Level
plt.figure(figsize=(14, 8))
experience_groups = pd.cut(X_test['Experience_Years'], 
                          bins=[0, 5, 10, 15, 20, 30, 50],
                          labels=['0-5', '6-10', '11-15', '16-20', '21-30', '30+'])
grouped_errors = pd.DataFrame({
    'Experience': experience_groups,
    'Error': residuals,
    'AbsoluteError': abs_errors
}).groupby('Experience').median()

plt.subplot(1, 2, 1)
sns.barplot(x=grouped_errors.index, y=grouped_errors['AbsoluteError'], 
            palette='viridis')
plt.title('Median Absolute Error by Experience Level', fontsize=14)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Median Absolute Error ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(alpha=0.3)

```

```python
plt.subplot(1, 2, 2)
sns.boxplot(x=experience_groups, y=abs_errors, palette='coolwarm')
plt.title('Error Distribution by Experience', fontsize=14)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Absolute Error ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('error_by_experience.png', dpi=300)
plt.show()
```

```python
# 4. Error by Salary Quantiles
plt.figure(figsize=(14, 8))
salary_quantiles = pd.qcut(y_test, q=10, labels=False)
quantile_errors = pd.DataFrame({
    'SalaryQuantile': salary_quantiles,
    'Error': residuals
}).groupby('SalaryQuantile').median()

plt.subplot(1, 2, 1)
sns.barplot(x=quantile_errors.index, y=quantile_errors['Error'], 
            palette='plasma')
plt.title('Median Prediction Error by Salary Range', fontsize=14)
plt.xlabel('Salary Decile (1=Lowest, 10=Highest)', fontsize=12)
plt.ylabel('Median Prediction Error ($)', fontsize=12)
plt.grid(alpha=0.3)

```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_test, y=residuals, hue=salary_quantiles, 
               palette='plasma', alpha=0.7, size=abs_errors, sizes=(20, 200))
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Error Distribution Across Salary Ranges', fontsize=14)
plt.xlabel('Actual Salary ($)', fontsize=12)
plt.ylabel('Prediction Error ($)', fontsize=12)
plt.grid(alpha=0.3)
plt.legend(title='Salary Decile')

plt.tight_layout()
plt.savefig('error_by_salary_range.png', dpi=300)
plt.show()

```
